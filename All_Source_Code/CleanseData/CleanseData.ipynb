```python
import numpy as np
z = np.array([[      1,      2],
              [ np.nan,      3],
              [      4, np.nan],
              [      5,      6]])
iz = np.any(np.isnan(z), axis=1)
print(~iz)
z = z[~iz]
print(z)
```

```python
import numpy as np
import pandas as pd
z = pd.DataFrame({'x':[1,np.nan,4,5],'y':[2,3,np.nan,6]})
print(z)
```

```python
result = z.dropna()
```

```python
result = z.fillna(z.mean())
```

```python
result = z[z['y']<5.5]
```

```python
result = z[(z['y']<5.5) & (z['y']>=1.0)]
```
